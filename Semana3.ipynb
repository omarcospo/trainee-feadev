## Exercício 1

Suponha que você é um vendedor de carros e quer organizar as vendas dos carros, para enteder o perfil de cada comprador. No entanto, está tudo bagunçado e de díficil vizualização. Para arrumar essa bagunça, você decidade usar pandas!

```python
import pandas as pd

df_carros = pd.DataFrame({
    'Carro':['versa', 'sentra', 'civic', 'corolla', 'versa', 'sentra', 'civic', 'corolla'],
    'Comprador':['Gabriel', 'Roberta', 'Pedro', 'Elza', 'Ana', 'Marcos', 'Julia', 'Adriano']
})
    
df_clientes = pd.DataFrame({
    'Comprador':['Gabriel', 'Roberta', 'Pedro', 'Elza', 'Ana', 'Marcos', 'Julia', 'Adriano'],
    'Idade':[25, 35, 27, 45, 26, 30, 32, 50  ]
})

print(df_carros)

print(df_clientes)
```

```python
df_conjunto = pd.merge(df_carros, df_clientes, on='Comprador')

print(df_conjunto)
```

```python
df_conjunto2 = df_conjunto.groupby(['Carro'])['Idade'].mean()

print(df_conjunto2)
```

Pessoas mais velhas preferem Corolla e mais jovens preferem Versa
